## Eşik değeri kullanarak fotoğraftan istenilen bölgeleri çıkarma

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from utils import gauss_kernel_olustur, konvolusyon
```

```python
foto = cv2.imread("./fotograflar/space.tif", 0)
print(foto.shape)
plt.imshow(foto, cmap="gray")
plt.show()
```

```python
esik_degeri = 125
esik_maske = foto > esik_degeri

plt.imshow(esik_maske, cmap="gray")
plt.show()
```

```python
sigma = 5
m = n = 6 * sigma + 1
gauss_kernel = gauss_kernel_olustur(m, n, K=1, sigma=sigma)
```

```python
cikis_fotograf = konvolusyon(foto, gauss_kernel)
```

```python
print(cikis_fotograf.shape)

plt.imshow(cikis_fotograf, cmap="gray")
plt.show()
```

```python
bulanik_esik_maske = cikis_fotograf > esik_degeri

plt.imshow(bulanik_esik_maske, cmap="gray")
plt.show()
```
