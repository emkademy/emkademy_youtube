# Fotoğraflardaki Şekillerin Konumlarını Yakalama

```python
def goster(foto, cmap="gray"):
    kopya = foto.copy()
    if len(kopya.shape) == 3:
        kopya = kopya[..., ::-1]
        cmap = None
    plt.figure(figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')
    plt.imshow(kopya, cmap=cmap)
    plt.show()
```

```python
foto = cv2.imread("./fotograflar/mixed_image.png")
goster(foto)
```

```python
siyah_beyaz = cv2.cvtColor(foto, cv2.COLOR_BGR2GRAY)
goster(siyah_beyaz)
```
